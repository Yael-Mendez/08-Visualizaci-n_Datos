2.Crea un grafico de barras en donde se muestren la cantidad de personas registradas por occupation.

```python
a=Es['occupation'].value_counts()
plt.figure(figsize=(10,6))  #tamaño de gráfico
plt.bar(a.index,a.values,color='red')
plt.title('Profesiones de los clientes')
plt.xlabel('Profeciones')#poner etiqueta a eje x
plt.ylabel('Número de Clientes')#poner etiqueta a eje y
plt.xticks(rotation=90)
plt.show()#mostrar Gráfico
```

3.Crea un gráfico de barras en donde se muestre el total de personar registradas por education_level

```python
a=Es['education_level'].value_counts()
plt.figure(figsize=(15,6))  #tamaño de gráfico
plt.bar(a.index,a.values,color='black')
plt.title('Nivel Educativo')
plt.xlabel('Profeciones')#poner etiqueta a eje x
plt.ylabel('Número de Clientes')#poner etiqueta a eje y
plt.xticks(rotation=90)
plt.show()#mostrar Gráfico
```

4.Crea un grafico circular en donde se muestre la proporcion de Hombres y mujeres de acuerdo a la columna gender

```python
a=Es['gender'].value_counts()
plt.figure(figsize=(15,6))  #tamaño de gráfico
plt.pie(a.values, labels=a.index, colors=['blue','pink'],autopct='%1f%%')
plt.title('Genero')
plt.show()#mostrar Gráfico
```

5. Crea un histograma con la columna age

```python
a=Es['age'].value_counts()
plt.figure(figsize=(15,6))  #tamaño de gráfico
plt.hist(Es['age'], bins=10,color='blue',edgecolor='black')

plt.title('Distribucion de Edad')
plt.xlabel('Edad')#poner etiqueta a eje x
plt.ylabel('Número de Clientes')#poner etiqueta a eje y
plt.xticks(rotation=90)
plt.show()#mostrar Gráfico
```
